```python
import pandas as pd
```

```python
tabla=pd.read_csv("C:/Users/claudio.pacheco/Downloads/prog_avance_de_indicadores.csv",encoding='latin-1', low_memory=False)
tabla.columns=tabla.columns.str.lower()
#filtra ciclo 2023
tabla=tabla[tabla['ciclo']==2023]
#Selecciona las columnas filtro, desc_ramo,id_objetivo
tabla=tabla[['ciclo','id_ramo','desc_ramo','id_modalidad','id_pp','desc_pp','id_objetivo','desc_objetivo', 'desc_nivel']]
tabla=tabla.drop_duplicates(subset=['id_objetivo'], keep='first')
tabla
```

```python
draft=tabla.copy()
draft=draft[draft["desc_nivel"]=="Fin"]
#Buscar si en desc_objetivo contiene la palabra mujer
draft["mujer"]=draft["desc_objetivo"].str.contains("mujer", case=False)
draft=draft.dropna(subset=['mujer'])
#Transformar a 1 y 0
draft["mujer"]=draft["mujer"].astype(int)
draft
```

```python
#Contar 1 en la columna de mujer
draft['mujer'].value_counts()
```

```python
tabla=pd.read_csv("C:/Users/claudio.pacheco/Downloads/prog_avance_de_indicadores.csv",encoding='latin-1', low_memory=False)
tabla.columns=tabla.columns.str.lower()
#filtra ciclo 2023
tabla=tabla[tabla['ciclo']==2023]
#Selecciona las columnas filtro, desc_ramo,id_objetivo
tabla=tabla[['ciclo','id_ramo','desc_ramo','id_modalidad','id_pp','desc_pp','id_objetivo','id_indicador','desc_objetivo', 'desc_indicador','desc_nivel']]
tabla=tabla.drop_duplicates(subset=['id_indicador'], keep='first')
#Dejar solo los indicadores a nivel componente
tabla=tabla[tabla['desc_nivel']=='Componente']
print("Total de indicadores a nivel componente: ",tabla.shape[0])
tabla["mujer"]=tabla["desc_indicador"].str.contains("mujer", case=False)
tabla=tabla.dropna(subset=['mujer'])
#Transformar a 1 y 0
tabla["mujer"]=tabla["mujer"].astype(int)
#Contar 1 en la columna de mujer
print(tabla['mujer'].value_counts())
tabla
```

```python
#Contar indicadores de mujer=1 por desc_ramo
print("Indicadores que contienen la palabra mujer por ramo")
tabla.groupby(['desc_ramo'])['mujer'].sum().sort_values(ascending=False)
```
